# Linear regression model of salary.csv dataset while splitting data for training and testing

```python
#importing required libraries
import pandas as pd
import numpy as np
import joblib
```

```python
#importing data
data = pd.read_csv("SalaryData.csv")
```

```python
#reviewing data through printing first 5 datasets
data.head()
```

```python
#assigning features and target variables
x = data["YearsExperience"].values
y = data["Salary"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.33, random_state = 0)
```

```python
X_train = X_train.reshape(20,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
X_test = X_test.reshape(10,1)
```

```python
model.predict(X_test)
```

```python
model.coef_
```

```python
joblib.dump(model, "mark2.pkl")
```
